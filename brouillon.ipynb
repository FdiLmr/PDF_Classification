```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('./data/cc-provenance-20230324-1k.csv')

# Get the column names
columns = df.columns

# Display the column names
print("Columns in the CSV file:")
for col in columns:
    print(col)
```

```python
import pandas as pd

# Read the classified PDFs CSV file
classified_df = pd.read_csv('./classified_pdfs_1k_gpt.csv')

# Count the occurrences of each classification
classification_counts = classified_df['classification'].value_counts()

# Display the counts
print("Classification counts:")
print(classification_counts)

# Optional: Create a bar plot of the counts
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
classification_counts.plot(kind='bar')
plt.title('Distribution of PDF Classifications')
plt.xlabel('Classification')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# Calculate the total sum of all classifications
total_classifications = classification_counts.sum()

# Print the total number of classifications
print(f"\nTotal number of classifications: {total_classifications}")


```
